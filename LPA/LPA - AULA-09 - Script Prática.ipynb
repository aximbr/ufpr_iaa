## Parte 4 - Dicionários

### Seção 1: Criação de dicionários

```python
# Dicionários - uso de chaves e separação de chaves e valores por dois pontos (:)
estudantes_dict = {"Mateus":24, "Fernanda":22, "Tamires":26, "Cristiano":25}
print(estudantes_dict)
```

```python
# Função type para retornar o tipo do objeto. Dict = dicionário
type(estudantes_dict)
```

```python
# Diferença para listas - uso de colchetes e elementos separados por vírgulas
estudantes_lst = ["Mateus", 24, "Fernanda", 22, "Tamires", 26, "Cristiano", 25] 
print(estudantes_lst)
```

```python
# Função type para retornar o tipo do objeto. List = lista
type(estudantes_lst)
```

```python
# Selecionando um elemento de um dicionário; retorno do valor 
estudantes_dict["Mateus"]
```

### Seção 2: Funções e métodos de dicionários

```python
# Excluindo todos os elementos de um dicionário - método clear
estudantes_dict.clear()
print(estudantes_dict)
```

```python
# Deletando a variável que contém os valores do dicionário- função del
del estudantes_dict
print(estudantes_dict)
```

```python
# Criando outro dicionário
nota_alunos = {"Breno":10, "Julia":5, "Manuela":7, "Ricardo":6}
```

```python
# Função len retorna o número de elementos do dicionário
len(nota_alunos)
```

```python
# Método keys retorna as chaves
nota_alunos.keys()
```

```python
# Método values retorna as valores
nota_alunos.values()
```

```python
# Método items retorno os valores em formato de lista
nota_alunos.items()
```

### Seção 3: Criando dicionários aninhados

```python
# Criando dicionários aninhados
dict_aninhado = {'key1':{'key2_aninhada':{'key3_aninhada':'Dict aninhado em Python'}}}
```

```python
print(dict_aninhado)
```

#### Nesse caso, recebendo o valor das chaves em uma variável "dict_aninhado". Note que para acessar o valor da terceira chave ("key_3") é necessário acessar os valores a partir de "key_1" e "key_2"

```python
dict_aninhado['key1']['key2_aninhada']['key3_aninhada']
```

## Parte 5 - Tuplas

### Seção 1: Criação de tuplas

```python
# Criando uma tupla
tupla1 = ("Geografia", 23, "Elefantes")
```

```python
# Imprimindo a tupla
print(tupla1)
```

```python
# Função type para retornar o tipo do objeto. Tuple = tupla
type(tupla1)
```

```python
# Tuplas não suportam o método append(), para adicionar elementos. Isso ocorre pois tuplas são imutáveis
tupla1.append("Chocolate")   
```

```python
# Tuplas não suportam a função del (delete) de um item específico
del tupla1["Geografia"]  
```

### Seção 2: Funções e métodos de tuplas

```python
# Verificando o comprimento da tupla - função len
len(tupla1)
```

```python
# Tuplas não suportam atribuição de item
tupla1[1] = 21
```

```python
# Deletando a tupla. Observe que a função del só vale para deletar o objeto que contém a tupla
del tupla1
```

```python
# recriando a tupla
tupla1 = ("Geografia", 23, "Elefantes")
```

```python
# Usando a função list() para converter uma tupla para lista
lista_tupla_1 = list(tupla1)
```

```python
print(lista_tupla_1)
```

```python
# Função type para retornar o tipo do objeto
type(lista_tupla_1)
```

```python
# Transformando uma tupla em lista, o objeto passa a ser mutável e aceitar métodos como append
lista_tupla_1.append('História')
```

```python
print(lista_tupla_1)
```

```python
# Em seguida, é possível converter o objeto novamente em tupla
# Essa manipulação e transformação de tipos de estruturas de dados é muito comum em Python
tupla_2 = tuple(lista_tupla_1)
```

```python
# Objeto tupla_2, agora com novo item adicionado (elemento "História")
print(tupla_2)
```
