# 1. Unpaired Z Test (amostras independentes)

```python
import numpy as np
import pandas as pd
from scipy import stats

# carregar dados
qi_unpaired = pd.read_csv("qi_unpaired.csv")

# separar grupos
boys = qi_unpaired.loc[qi_unpaired["group"] == "boy", "qi"].values
girls = qi_unpaired.loc[qi_unpaired["group"] == "girl", "qi"].values

# -------------------------
# Teste de normalidade
# -------------------------
print("Shapiro boys:", stats.shapiro(boys))
print("Shapiro girls:", stats.shapiro(girls))

# -------------------------
# Teste F (homogeneidade)
# -------------------------
var_boys = np.var(boys, ddof=1)
var_girls = np.var(girls, ddof=1)

F = var_boys / var_girls
df1 = len(boys) - 1
df2 = len(girls) - 1

p_value_f = 2 * min(
    stats.f.cdf(F, df1, df2),
    1 - stats.f.cdf(F, df1, df2)
)

print("F =", F)
print("p-value =", p_value_f)

# -------------------------
# Z-test duas amostras
# -------------------------
mean_boys = np.mean(boys)
mean_girls = np.mean(girls)

sigma_boys = np.std(boys, ddof=1)
sigma_girls = np.std(girls, ddof=1)

n1 = len(boys)
n2 = len(girls)

z = (mean_boys - mean_girls) / np.sqrt(
    (sigma_boys**2 / n1) + (sigma_girls**2 / n2)
)

p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("Z =", z)
print("p-value =", p_value)

# intervalo de confiança
z_crit = stats.norm.ppf(0.975)

diff = mean_boys - mean_girls
se = np.sqrt((sigma_boys**2/n1)+(sigma_girls**2/n2))

ci_low = diff - z_crit*se
ci_high = diff + z_crit*se

print("IC95%:", ci_low, ci_high)
```

# 2. Unpaired t Test (amostras independentes)

```python
import numpy as np
import pandas as pd
from scipy import stats

mw_weight = pd.read_csv("mw_weight.csv")

men = mw_weight.loc[mw_weight["group"]=="Man","weight"].values
women = mw_weight.loc[mw_weight["group"]=="Woman","weight"].values

# -------------------------
# normalidade
# -------------------------
print("Shapiro men:", stats.shapiro(men))
print("Shapiro women:", stats.shapiro(women))

# -------------------------
# F-test variâncias
# -------------------------
var_men = np.var(men, ddof=1)
var_women = np.var(women, ddof=1)

F = var_men/var_women
df1 = len(men)-1
df2 = len(women)-1

p_value_f = 2*min(
    stats.f.cdf(F,df1,df2),
    1-stats.f.cdf(F,df1,df2)
)

print("F =",F,"p=",p_value_f)

# -------------------------
# t-test variância igual
# -------------------------
t_stat, p_value = stats.ttest_ind(
    men,
    women,
    equal_var=True
)

print("t =", t_stat)
print("p =", p_value)

# intervalo de confiança
diff = np.mean(men) - np.mean(women)

n1 = len(men)
n2 = len(women)

sp = np.sqrt(
    ((n1-1)*np.var(men,ddof=1)+(n2-1)*np.var(women,ddof=1)) /
    (n1+n2-2)
)

se = sp*np.sqrt(1/n1 + 1/n2)

t_crit = stats.t.ppf(0.975, n1+n2-2)

ci_low = diff - t_crit*se
ci_high = diff + t_crit*se

print("IC95%:", ci_low, ci_high)
```
