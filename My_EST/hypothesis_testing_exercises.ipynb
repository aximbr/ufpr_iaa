## 1. Setup

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

## 2. Load the synthetic dataset

```python
df = pd.read_csv("synthetic_hypothesis_dataset.csv")
df.head()
```

Dataset columns:

- group: categorical group (A, B, C)
- before_score: score before training
- after_score: score after training
- algorithm: algorithm identifier
- accuracy: simulated model accuracy

## 4. Test for normality (Shapiro–Wilk)

## 5. One‑sample t‑test


Hypothesis example:

H0: mean(before_score) = 70


```python
stats.ttest_1samp(df["before_score"], 70)
```

Interpret the p‑value.

## 6. Paired t‑test (before vs after training)

```python
stats.ttest_rel(df["before_score"], df["after_score"])
```

Exercise:

Does the training significantly change the scores?


## 7. Independent t‑test (algorithm accuracy)

```python
alg1 = df[df.algorithm=="alg1"]["accuracy"]
alg2 = df[df.algorithm=="alg2"]["accuracy"]

stats.ttest_ind(alg1, alg2)
```

Interpret whether the algorithms differ.

## 8. Non‑parametric alternative (Mann‑Whitney U)

```python
stats.mannwhitneyu(alg1, alg2)
```

Compare this result with the t‑test.


## 9. ANOVA (3 groups)

```python
g1 = df[df.group=="A"]["before_score"]
g2 = df[df.group=="B"]["before_score"]
g3 = df[df.group=="C"]["before_score"]

stats.f_oneway(g1, g2, g3)
```

Exercise:

## 10. Bootstrap estimation

```python
data = df["accuracy"].values

boot_means = []
for _ in range(5000):
    sample = np.random.choice(data, size=len(data), replace=True)
    boot_means.append(sample.mean())

np.percentile(boot_means, [2.5, 97.5])
```

Interpret the bootstrap confidence interval.
